```python
# TASK 4.1.2

import numpy as np

print("1. Array with values from 10 to 49:")
arr1 = np.arange(10, 50)
print(arr1, "\n")

print("2. 3x3 matrix with values 0 to 8:")
matrix = np.arange(9).reshape(3, 3)
print(matrix, "\n")

print("3. 3x3 identity matrix:")
identity_matrix = np.eye(3)
print(identity_matrix, "\n")

print("4. Random array of size 30 and its mean:")
rand_arr = np.random.random(30)
print(rand_arr)
print("Mean:", rand_arr.mean(), "\n")

print("5. 10x10 random array with min & max values:")
random = np.random.random((10, 10))
print(random)
print("Min:", random.min(), "Max:", random.max(), "\n")

print("6. Zero array of size 10 with 5th element as 1:")
zero_arr = np.zeros(10)
zero_arr[4] = 1
print(zero_arr, "\n")

print("7. Reversed array [1,2,0,0,4,0]:")
arr = np.array([1, 2, 0, 0, 4, 0])
reversed_arr = arr[::-1]
print(reversed_arr, "\n")

print("8. 2D array with 1 on border and 0 inside (5x5):")
border_arr = np.ones((5, 5))
border_arr[1:-1, 1:-1] = 0
print(border_arr, "\n")

print("9. 8x8 Checkerboard pattern:")
checkerboard = np.zeros((8, 8))
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print(checkerboard)
```

```python
# TASK 4.1.3

x = np.array([[1, 2],
              [3, 5]])

y = np.array([[5, 6],
              [7, 8]])

v = np.array([9, 10])
w = np.array([11, 12])

print("1. Add the two arrays (x + y):")
print(x + y, "\n")

print("2. Subtract the two arrays (x - y):")
print(x - y, "\n")

print("3. Multiply array x with any integer:")
print(x * 3, "\n")

print("4. Square of each element of array x:")
print(x ** 2, "\n")

print("5. Dot products:")

print("Dot product v · w:")
print(np.dot(v, w))
print()

print("Dot product x · v:")
print(np.dot(x, v))
print()

print("Dot product x · y:")
print(np.dot(x, y))
print()

print("6. Concatenate x and y along rows:")
concat_xy = np.concatenate((x, y), axis=0)
print(concat_xy, "\n")

print("Concatenate v and w along columns:")
concat_vw = np.vstack((v, w))
print(concat_vw, "\n")

print("7. Try to concatenate x and v:")
try:
    result = np.concatenate((x, v), axis=0)
    print(result)
except Exception as e:
    print("Error:", e)
    print("\nExplanation:")
    print("""
    x is a 2×2 matrix (2 rows, 2 columns)
    v is a 1D array with shape (2,)

    When concatenating along rows (axis=0), NumPy expects both arrays
    to have the same number of columns.

    But:
        x has shape (2, 2)
        v has shape (2,) → treated as (2,) not (1,2)

    Because v does NOT have 2 dimensions, NumPy cannot stack it with x directly.
    """)

```

```python
# TASK 4.1.4
# Part 1: Matrices A and B

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print("A * A^-1 =\n", np.round(I, 2))

# Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nAB =\n", AB)
print("BA =\n", BA)

# Prove (AB)ᵀ = Bᵀ * Aᵀ
LHS = AB.T
RHS = np.dot(B.T, A.T)
print("\n\n(AB)^T =\n", LHS)
print("B^T * A^T =\n", RHS)


# Part 2: Solve linear system
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

A_sys = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3, 1, -1]])
B_sys = np.array([-1, -3, 9])

# Method 1: Using inverse
X_inv = np.dot(np.linalg.inv(A_sys), B_sys)
print("\nSolution using inverse method [x, y, z] =", np.round(X_inv, 2))

# Method 2: Using np.linalg.solve (more efficient)
X_solve = np.linalg.solve(A_sys, B_sys)
print("Solution using np.linalg.solve [x, y, z] =", np.round(X_solve, 2))
```

```python
# TASK 4.2.1

import time
import numpy as np

print("1. ELEMENT-WISE ADDITION")

# Python list addition
list1 = list(range(1_000_000))
list2 = list(range(1_000_000))

start = time.time()
list_add = [list1[i] + list2[i] for i in range(1_000_000)]
end = time.time()
print("Python List Addition Time:", end - start, "seconds")

# NumPy addition
arr1 = np.arange(1_000_000)
arr2 = np.arange(1_000_000)

start = time.time()
arr_add = arr1 + arr2
end = time.time()
print("NumPy Array Addition Time:", end - start, "seconds")


print("\n2. ELEMENT-WISE MULTIPLICATION")

# Python list multiplication
start = time.time()
list_mul = [list1[i] * list2[i] for i in range(1_000_000)]
end = time.time()
print("Python List Multiplication Time:", end - start, "seconds")

# NumPy multiplication
start = time.time()
arr_mul = arr1 * arr2
end = time.time()
print("NumPy Array Multiplication Time:", end - start, "seconds")


print("\n3. DOT PRODUCT")

# Python list dot product
start = time.time()
dot_py = sum(list1[i] * list2[i] for i in range(1_000_000))
end = time.time()
print("Python List Dot Product Time:", end - start, "seconds")

# NumPy dot product
start = time.time()
dot_np = np.dot(arr1, arr2)
end = time.time()
print("NumPy Dot Product Time:", end - start, "seconds")


print("\n4. MATRIX MULTIPLICATION (1000 x 1000)")

# Python list matrix multiplication
matrix1 = [[1]*1000 for _ in range(1000)]
matrix2 = [[1]*1000 for _ in range(1000)]

start = time.time()
result_matrix = [
    [sum(matrix1[i][k] * matrix2[k][j] for k in range(1000))
    for j in range(1000)]
    for i in range(1000)
]
end = time.time()
print("Python List Matrix Multiplication Time:", end - start, "seconds")

# NumPy matrix multiplication
np_matrix1 = np.ones((1000, 1000))
np_matrix2 = np.ones((1000, 1000))

start = time.time()
np_result = np.dot(np_matrix1, np_matrix2)
end = time.time()
print("NumPy Matrix Multiplication Time:", end - start, "seconds")
```
